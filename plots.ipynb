```python
import plotly.graph_objects as go
import pandas as pd
```

```python
pendulum_0001 = (pd.read_csv('scores_tau_0.001.csv')
                 .assign(tau=0.001,
                         env = 'Pendulum-v0'))
pendulum_001 = (pd.read_csv('scores_tau_0.01.csv')
                 .assign(tau=0.01,
                         env = 'Pendulum-v0')
                 )
pendulum_01 = (pd.read_csv('scores_tau_0.1.csv')
                 .assign(tau=0.1,
                         env = 'Pendulum-v0')
                 )

moutaincar_0001 = (pd.read_csv('scores_tau_0.001_scale_1_MountainCarContinuous-v0.csv')
                    .assign(tau=0.001,
                            env = 'MountainCarContinuous-v0'))


moutaincar_001 = (pd.read_csv('scores_tau_0.01_MountainCarContinuous-v0.csv')
                    .assign(tau=0.01,
                            env = 'MountainCarContinuous-v0'))

moutaincar_01 = (pd.read_csv('scores_tau_0.1_MountainCarContinuous-v0.csv')
                    .assign(tau=0.1,
                            env = 'MountainCarContinuous-v0'))
```

```python
all_scores = pd.concat([pendulum_0001, pendulum_001, pendulum_01, moutaincar_0001, moutaincar_01, moutaincar_001]).assign(epochs = lambda df : df.index)

pendulum = all_scores.query('env == "Pendulum-v0"')
mountaincar = all_scores.query('env == "MountainCarContinuous-v0"')

# smooth scores for every tau
pendulum = pendulum.groupby(['tau','epochs']).Scores.mean().reset_index()
# mountaincar = mountaincar.groupby('tau').mean().reset_index()


fig = go.Figure()

# plot pendulum scores according to tau
for tau in pendulum['tau'].unique():
    data = pendulum.query(f'tau == {tau}')
    fig.add_trace(go.Scatter(x=data.epochs, y=data['Scores'], mode='lines', name=f'tau={tau}'))
    
fig.update_layout(title='Pendulum-v0 scores according to tau',
                    xaxis_title='Epochs',
                    yaxis_title='Scores',
                    width=600,
                    height=600)




                    
                    
```

```python
mountaincar = all_scores.query('env == "MountainCarContinuous-v0" and tau != 0.001')



fig = go.Figure()

# plot pendulum scores according to tau
for tau in mountaincar['tau'].unique():
    data = mountaincar.query(f'tau == {tau}')
    fig.add_trace(go.Scatter(x=data.index, y=data['Scores'], mode='lines', name=f'tau={tau}'))
    
fig.update_layout(title='MountainCarContinuous-v0 scores according to tau',
                    xaxis_title='Epochs',
                    yaxis_title='Scores',
                    width=600,
                    height=600)
```

### Different scales

```python
pendulum_scale_1 = (pd.read_csv('scores_tau_0.1.csv')
                    .assign(scale = 1,
                            env = 'Pendulum-v0'))
                    
pendulum_scale_5 = (pd.read_csv('scores_tau_0.1_scale_5_Pendulum-v1.csv')
                    .assign(scale = 5,
                            env = 'Pendulum-v0'))

pendulum_scale_10 = (pd.read_csv('scores_tau_0.1_scale_10_Pendulum-v1.csv')
                    .assign(scale = 10,
                            env = 'Pendulum-v0'))


all_scales = pd.concat([pendulum_scale_1, pendulum_scale_5, pendulum_scale_10]).assign(epochs = lambda df : df.index)
```

```python
# plot scores according to scale

fig = go.Figure()

for scale in all_scales['scale'].unique():
    data = all_scales.query(f'scale == {scale}')
    fig.add_trace(go.Scatter
                    (x=data.epochs, y=data['Scores'], mode='lines', name=f'scale={scale}'))
    
fig.update_layout(title='Pendulum-v0 scores according to scale',
                    xaxis_title='Epochs',
                    yaxis_title='Scores',
                    width=1200,
                    height=600)
```
